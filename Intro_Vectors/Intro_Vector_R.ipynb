5. Let **u** = (1,2,-2), **v** = (3,-12,4), and a scalar *k* = -3 <br>
    Find ||u||, ||v||, ||u+v||, ||*k*u|| <br>

    $||u|| = \sqrt{u \cdot u} = \sqrt{1^2 + 2^2 + (-2)^2} = \sqrt{1 + 4 + 4} = \sqrt{9} = 3 $<br><br>
    $||v|| = \sqrt{v \cdot v} = \sqrt{3^2 + (-12)^2 + 4^2} = \sqrt{9 + 144 + 16} = \sqrt{169} = 13 $<br><br>
    $||u+v|| = \sqrt{(u+v)^2} = \sqrt{(1+3)^2 + (2+(-12))^2 + (-2+4)^2} = \sqrt{16 + 100 + 4} = \sqrt{120} $<br><br>
    $||*k*u|| = \sqrt{(k \cdot u)^2} = \sqrt{(1(-3))^2 + (2(-3))^2 + (-2(-3))^2} = \sqrt{9 + 36 + 36} = \sqrt{81} = 9 $<br><br>

```python
import numpy as np
u = ([1,2,-2])
v = ([3,-12,4])
k = -3

print("||u|| = ", np.linalg.norm(u))
print("||v|| = ", np.linalg.norm(v))
print("||u+v|| = ", np.linalg.norm(u+v))
print("||ku|| = ", np.linalg.norm(u)*k)

```
